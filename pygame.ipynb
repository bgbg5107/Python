```python
import pygame
import random
import time

screen_width = 480
screen_height = 700

pygame.init()

screen = pygame.display.set_mode((screen_width,screen_height))
pygame.display.set_caption("해피할로윈데이")

clock = pygame.time.Clock()
total_time = 10000 # 제한시간
# gold_time = 10000

# counter = 5
chance = 5
score = 0

# 배경만들기
background = pygame.image.load("/Users/jiyoung/background.png")
    
# 캐릭터만들기
character = pygame.image.load("/Users/jiyoung/ghost.png")
character_size = character.get_rect().size
character_width = character_size[0]
character_height = character_size[1]
character_x_pos = (screen_width - character_width) / 2
character_y_pos = screen_height - character_height

to_x = 0  # 이동위치 변수 생성
speed = 0.8

# candyList = [candy1, candy2, gold]

# 먹어야하는 것(캔디) 만들기
candy1 = pygame.image.load("/Users/jiyoung/candy.png")
candy1_size = candy1.get_rect().size
candy1_width = candy1_size[0]
candy1_height = candy1_size[1]
candy1_x_pos = random.randint(0, screen_width - candy1_width)
candy1_y_pos = 0
candy1_speed = 35

# 먹어야하는 것(캔디2) 만들기
candy2 = pygame.image.load("/Users/jiyoung/candy2.png")
candy2_size = candy2.get_rect().size
candy2_width = candy2_size[0]
candy2_height = candy2_size[1]
candy2_x_pos = random.randint(0, screen_width - candy2_width)
candy2_y_pos = 0
candy2_speed = 30

# 먹어야하는 것(gold candy) 만들기
gold = pygame.image.load("/Users/jiyoung/gold.png")
gold_size = gold.get_rect().size
gold_width = gold_size[0]
gold_height = gold_size[1]
gold_x_pos = random.randint(0, screen_width - gold_width)
gold_y_pos = -random.randint(gold_height, 5000) #x축 대칭해서 떨어지는 높이 다르게 설정
# gold_y_pos = 0
gold_speed = 30

# 피해야할 적(똥) 만들기2
pumpkin1 = pygame.image.load("/Users/jiyoung/pumpkin.png")
pumpkin1_size = pumpkin1.get_rect().size
pumpkin1_width = pumpkin1_size[0]
pumpkin1_height = pumpkin1_size[1]
pumpkin1_x_pos = random.randint(0, screen_width - pumpkin1_width)
pumpkin1_y_pos = 0
pumpkin1_speed = 20

game_font = pygame.font.SysFont(None, 40)

start_ticks = pygame.time.get_ticks()


##게임 시작화면
running = False
while running == False:
    clock.tick(30)
    for event in pygame.event.get():
        if(event.type == pygame.KEYDOWN):
            if(event.key == pygame.K_SPACE):
                running = True
    game_font = pygame.font.SysFont(None, 40)
    text = game_font.render("HAPPY HALLOWEEN", True,(255, 0, 0))
    text2 = game_font.render("PRESS SPACE KEY", True,(255, 0, 0))
    screen.blit(text, (100,300))
    screen.blit(text2, (100,340))
    pygame.display.update()
    
while running:
    dt = clock.tick(30)
    # 2. 이벤트 처리(키보드, 마우스 등)
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                to_x -= speed
            elif event.key == pygame.K_RIGHT:
                to_x += speed
#             elif event.key == pygame.K_ENTER:  게임 중지하려면 어떻게...?
#                 time.sleep(5)

        if event.type == pygame.KEYUP:
            if event.key == pygame.K_LEFT or event.key == pygame.K_RIGHT:
                to_x = 0

    # 3. 게임캐릭터 위치정의
    character_x_pos += to_x * dt
    
    # 떨어지기 설정
    candy1_y_pos += candy1_speed 
    candy2_y_pos += candy2_speed
    gold_y_pos += gold_speed
    pumpkin1_y_pos += pumpkin1_speed

    if candy1_y_pos > screen_height:
        candy1_y_pos = 0
        candy1_x_pos = random.randint(0, screen_width - candy1_width)
        
    if candy2_y_pos > screen_height:
        candy2_y_pos = 0
        candy2_x_pos = random.randint(0, screen_width - candy2_width)
        
    if gold_y_pos > screen_height:
        gold_y_pos = -5000
        gold_x_pos = random.randint(0, screen_width - gold_width)
        
    if pumpkin1_y_pos > screen_height:
        pumpkin1_y_pos = 0
        pumpkin1_x_pos = random.randint(0, screen_width - pumpkin1_width)

    if character_x_pos < 0:
        character_x_pos = 0
    elif character_x_pos > screen_width - character_width:
        character_x_pos = screen_width - character_width

    # 4. 충돌 처리
    character_rect = character.get_rect()
    character_rect.left = character_x_pos
    character_rect.top = character_y_pos

    candy1_rect = candy1.get_rect()
    candy1_rect.left = candy1_x_pos
    candy1_rect.top = candy1_y_pos
    
    candy2_rect = candy2.get_rect()
    candy2_rect.left = candy2_x_pos
    candy2_rect.top = candy2_y_pos
    
    gold_rect = gold.get_rect()
    gold_rect.left = gold_x_pos
    gold_rect.top = gold_y_pos
    
    pumpkin1_rect = pumpkin1.get_rect()
    pumpkin1_rect.left = pumpkin1_x_pos
    pumpkin1_rect.top = pumpkin1_y_pos

    if character_rect.colliderect(candy1_rect): #candy
        candy1_y_pos = 0
        candy1_x_pos = random.randint(0, screen_width - candy1_width)
#         print("캔디획득!")
        score = score + 1
        running = True
        
    if character_rect.colliderect(candy2_rect): #candy2
        candy2_y_pos = 0
        candy2_x_pos = random.randint(0, screen_width - candy2_width)
#         print("캔디2획득!")
        score = score + 1
        running = True
    ###################################################################################################
    if character_rect.colliderect(gold_rect): #gold candy
#         candy1 = gold
#         candy2 = gold
        pygame.display.update()
        gold_y_pos = 0
        gold_x_pos = random.randint(0, screen_width - gold_width)
#         score = score + 1 #안되면 그냥 골드캔드 먹으면 +5???
        running = True
        
    if character_rect.colliderect(pumpkin1_rect):
        pumpkin1_y_pos = 0
        pumpkin1_x_pos = random.randint(0, screen_width - pumpkin1_width)
#         print("호박충돌")
        chance = chance - 1
        if(chance == 0):
            running = False
            while running == False:
                clock.tick(30)
                time.sleep(1)
                for event in pygame.event.get():
                    if(event.type == pygame.KEYDOWN):
                        if(event.key == pygame.K_SPACE):
                            running = True
                        if(event.type == pygame.QUIT):
                            running = False
                game_font = pygame.font.SysFont(None, 40)
                text = game_font.render("GAME OVER", True,(255, 0, 0))
                screen.blit(text, (150,350))
                pygame.display.update()
            
            chance = 5
            score = 0
            character = pygame.image.load("/Users/jiyoung/ghost.png")
    # 5. 화면에 그리기
    screen.blit(background, (0, 0))
    screen.blit(character, (character_x_pos, character_y_pos))
    screen.blit(candy1, (candy1_x_pos, candy1_y_pos))
    screen.blit(candy2, (candy2_x_pos, candy2_y_pos))
    screen.blit(gold, (gold_x_pos, gold_y_pos))
    screen.blit(pumpkin1, (pumpkin1_x_pos, pumpkin1_y_pos))

    elapsed_time = (pygame.time.get_ticks() - start_ticks / 1000)

#     timer = game_font.render(str(int(total_time - elapsed_time)), True,(0, 0, 0))
    text = game_font.render("Chance:{} score:{}".format(chance,score), True,(255, 0, 0))
#     timer = game_font.render(str(int(gold_time - elapsed_time)), True,(0, 0, 0))
#     count = game_font.render(str(counter), True,(0, 0, 0))
#     screen.blit(count, (10, 20))
#     screen.blit(timer, (10, 60))
    screen.blit(text, (15,40))

#     if total_time - elapsed_time <= 0:
#     if gold_time - elapsed_time <= 0:
#         print("골드타임아웃")
#         running = False

    pygame.display.update()
pygame.quit()
```
