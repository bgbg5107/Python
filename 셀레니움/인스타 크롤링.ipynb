```python
# 첫번째 게시물 클릭하기
first_article = browser.find_elements_by_css_selector('div._aagw')[0]
first_article.click()
results = []
# 페이지 소스 가져오기
for i in range(50):
    html = browser.page_source
    soup = BeautifulSoup(html, 'html.parser')

# 본문 찾기
    try:
        content = soup.select('div._a9zs')[0].text
    except:
        content = ''
    
# 좋아요 찾기
    try:
        likes = soup.select('div._aacl._aaco._aacw._adda._aacx._aada._aade')[0].text
    except: # 동영상이거나, 좋아요가 없거나..
        likes = 0
    try:
        view = soup.select('section._aat4>div>span>div>span')[0].text
    except: # 사진이거나 조회수가 없거나..?
        view = 0
    try: # 가장 먼저 좋아요를 눌러보세요
        likes2 = soup.select('div._aacl._aaco._aacu._aacx._aad6._aade')[0].text
    except:
        likes2 = 0
       
    date = soup.select('div._aacl._aacm._aacu._aacy._aad6>time')[0]['datetime'][0:10]
    
    print(content, likes, view, date)
    
# 우측 화살표 클릭하기
    next_button = browser.find_elements_by_css_selector('div._aaqg._aaqh')[0]
    next_button.click()
    time.sleep(1)

# 데이터프레임에 저장하기
    data = [content,likes,view,date]
    results.append(data)
```
